```python
import pandas as pd
```

```python
df = pd.read_csv("file7.csv", sep="\t")
df
```

Using pandas to create dummy variables

```python
#returns the dummy variable columns.
#for the town column, it will return the dummies for the 3 categories but in binaries
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged = pd.concat([df,dummies],axis='columns')
merged
```

```python
final = merged.drop(['town'], axis='columns')
final   #dropping the town column with the word rows.
```

Dummy Variable Trap
When you can derive one variable from other variables, they are known to be multi-colinear. Here if you know values of california and georgia then you can easily infer value of new jersey state, i.e. california=0 and georgia=0. There for these state variables are called to be multi-colinear. In this situation linear regression won't work as expected. Hence you need to drop one column.

NOTE: sklearn library takes care of dummy variable trap hence even if you don't drop one of the state columns it is going to work, however we should make a habit of taking care of dummy variable trap ourselves just in case library that you are using is not handling this for you

```python
final = final.drop(['west windsor'], axis='columns')  #dropping the west windsor
final
```

```python
#Dropping the price column since it is the dependent variable
X = final.drop('price', axis='columns')
X
```

```python
y = final.price
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.predict(X)
```

```python
#Calculates all the predicted values of all the rows in X and compares the predicted value with the actual value,
#which is y.After that it will use some formula to calculate the score.so the score here is 95%

model.score(X,y) #Using score method after building the model to find its accuracy like 95% accuracy.
```

```python
model.predict([[3400,0,0]]) # 3400 sqr ft home in west windsor, we make it 0,0 because west windsor stands in the 
#third position which will be 1 but not shown here.
```

```python
model.predict([[2800,0,1]]) # 2800 sqr ft home in robbinsville, we are predicting for robbinsville,so we make the second
#position 1 according to the positions
```

# Using sklearn OneHotEncoder

```python
df
```

```python
# First step is to use label encoder to convert town names into numbers
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()  #making label encoder the class object
```

```python
dfle = df #using the label encoder object on the original data frame, and making our new df as dfle
dfle.town = le.fit_transform(dfle.town) #fit transform means,it takes the label column as an input and returns the lael
dfle
```

```python
#getting our X variables.
X = dfle[['town','area']].values
X
```

```python
y = dfle.price.values
y
```

```python
#Now use one hot encoder to create dummy variables for each of the town
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town', OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
X = ct.fit_transform(X)
X
```

```python
X = X[:,1:] #meaning, take all the rows and for the columns take from 1 and onwards.
#so, it drops the zero column.
```

```python
X
```

```python
model.fit(X,y)
```

```python
model.predict([[0,1,3400]]) # 3400 sqr ft home in west windsor
#take note of the order of columns
```
